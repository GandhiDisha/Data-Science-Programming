##### <i>Question 1: Rock, Paper, Scissors</i>

```python
import random

def computer_brains(human_choice_count):
    """
    Returns the choice that the computer will make looking at the past player inputs.
    The choice made by human the maximum times is considered. If the max choice selected is just 1, then the choice by computer is obvious.
    But if it's two then the computer will be choosing the one which will either reult in a draw or the computer winning with the assumption that the player plays their max choice.
    And if all three choices have occured the same number of times then a random choice is taken into consideration.
    
    Parameters
    ----------
    human_choice_count : dict
        dictionary having key as all the options a player can choose with the value being the count of those choices.

    Returns
    -------
    string
        The choice that the computer will make.
        
    Example
    --------
    >>> computer_brains({'rock':3,'paper':2,'scissor':1})
    paper
    """
    
    max_human_choice_list = [k for k, v in human_choice_count.items() if v == max(human_choice_count.values())]

    if len(max_human_choice_list) == 2:
        if 'rock' in max_human_choice_list and 'paper' in max_human_choice_list:
            max_human_choice = 'rock'
        elif 'rock' in max_human_choice_list and 'scissor' in max_human_choice_list:
            max_human_choice = 'scissor'
        else: max_human_choice = 'paper'
    else:
        max_human_choice = random.choice(max_human_choice_list)
        
    if max_human_choice == 'rock':
        computer_choice = 'paper'
    elif max_human_choice == 'paper':
        computer_choice = 'scissor'
    else: computer_choice = 'rock'

    return computer_choice

def game_logic(human_selection, computer_selection):
    """
    Returns the winner depending on what the computer chose and what the human chose.
    The winner is decided based on the simple game rules given in the question and along with that the human choice is returned for saving it for decision making.
    
    Parameters
    ----------
    human_selection : string
        a string saying what did the human chose from rock, paper, or scissor.
    computer_selection : string
        a string saying what did we decide for the computer from rock, paper, or scissor.
        
    Returns
    -------
    string
        The winner as decided by the game rules
    string
        The choice made by human just for processing
        
    Example
    --------
    >>> game_logic('rock','paper')
    paper,rock
    """
    
    if human_selection == 'rock' and computer_selection == 'paper':
        winner = 'computer'
        choice = human_selection
    elif human_selection == 'rock' and computer_selection == 'scissor':
        winner = 'human'
        choice = human_selection
    elif human_selection == 'paper' and computer_selection == 'rock':
        winner = 'human'
        choice = human_selection
    elif human_selection == 'paper' and computer_selection == 'scissor':
        winner = 'computer'
        choice = human_selection
    elif human_selection == 'scissor' and computer_selection == 'rock':
        winner = 'computer'
        choice = human_selection
    elif human_selection == 'scissor' and computer_selection == 'paper':
        winner = 'human'
        choice = human_selection
    else: 
        winner = 'draw'
        choice = human_selection
    
    return winner, choice

def main():
    
    computer_choices = ['rock','paper','scissor']
    human_choices = ['rock','paper','scissor','quit']
    no_of_draws = 0
    no_of_total_games = 0
    no_of_human_wins = 0
    human_choice_count = dict((element,0) for element in computer_choices)
    
    while True:
    
        human_selection = str(input("Please Play Rock, Paper, Scissor, or Quit!  "))

        #validating user choice
        if human_selection in human_choices:

            #computer using their brains to choose their selection
            if no_of_total_games == 0:
                computer_selection = random.choice(computer_choices)
            else: 
                computer_selection = computer_brains(human_choice_count)

            if human_selection != 'quit':
                no_of_total_games = no_of_total_games + 1
                winner, choice = game_logic(human_selection, computer_selection)
                print("The Computer has chosen " + computer_selection + '!')
                
                #managing choice count
                if choice == 'rock':
                    human_choice_count['rock'] = human_choice_count['rock'] + 1
                elif choice == 'paper':
                    human_choice_count['paper'] = human_choice_count['paper'] + 1
                else: 
                    human_choice_count['scissor'] = human_choice_count['scissor'] + 1

                #managing winner count
                if winner == 'human':
                    no_of_human_wins = no_of_human_wins + 1
                    print("Congratulations, you win!")
                elif winner == 'draw':
                    no_of_draws = no_of_draws + 1
                    print("Oopsie, it's a draw!")
                else: 
                    print("Oh no, I win! B)")

            else:
                print("No of Games Played: " + str(no_of_total_games))
                print("No of Wins by the Player: " + str(no_of_human_wins))
                break

        else:
            print("Please input a valid choice in between rock, paper, scissor or quit")
    
if __name__ == "__main__":
    main()
```

 ##### <i>Question 3: Near Duplicate Detection</i>

```python
import itertools
import math

#3a to split the tweet into phrases of 3 words each with the output being a dictionary
def moving_window(tweet):
    """
    Splits a tweet into a dictionary containing all possible combinations of 3 word phrases in the continuous order as keys and 1 as their respective values.
    First the string is split into all the words and then the combination of phrases are created.
    Followed by appending them to dictionary keys.
    
    Parameters
    ----------
    tweet : string
        a whole tweet as a string
        
    Returns
    -------
    dict
        Dictionary of 3 word phrases as gathered from the tweet.
        
    Example
    --------
    >>> moving_window('This is Data Science homework')
    {'This is Data':1,'is Data Science':1,'Data Science homework':1}
    """
    split_string = str(tweet).split()
    phrases = [" ".join(split_string[i:i+3]) for i in range(len(split_string) - 2)]
    dict_of_tweets = {phrase: '1' for phrase in phrases}
    return dict_of_tweets

#3b to check the number of matching phrases of 3 words inside both the dictionaries and return the cosine similarity figure
def cosine(tweet1,tweet2):
    """
    This function finds out the matching phrases in between two tweets by comparing their phrase dictionary.
    All the matching phrases are added to a list named matching_phrase.
    Then cosine similarity is calculated using the formula given in the question.
    
    Parameters
    ----------
    tweet1 : dict
        dictionary of 3 word phrases as gathered from the first tweet.
        
    tweet2 : dict
        dictionary of 3 word phrases as gathered from the second tweet.
        
    Returns
    -------
    float
        cosine similarity as calculated with the formula no.of.phrases/len(tweet1)*len(tweet2)
        
    Example
    --------
    >>> cosine({'This is Data':1,'is Data Science':1,'Data Science homework':1},{'This is Data':1,'for Data Science':1,'Data Science homework':1})
    0.67
    """
    matching_phrase = []
    tweet1 = set(tweet1)
    tweet2 = set(tweet2)

    for tweet in tweet1.intersection(tweet2):
        matching_phrase.append(tweet)
        
    no_of_common_phrases = len(matching_phrase)
    cosine_similarity = no_of_common_phrases/math.sqrt(len(tweet1)*len(tweet2))
    
    return cosine_similarity

#3c to check the similarity in all possible unique combinations of 2 tweets
def main():
    
    #reading the file
    with open("data/Santa.txt") as file:
        tweets = file.readlines()
    
    #checking for similarity in between tweets
    similarity_counter =0
    for a, b in itertools.combinations(tweets, 2):
        tweet1_dict = moving_window(a)
        tweet2_dict = moving_window(b)
        cosine_similarity = cosine(tweet1_dict,tweet2_dict)
        if cosine_similarity > 0.5:
            print ("TWEET " + a + " and TWEET " + b + " are near similar!")
            similarity_counter = similarity_counter + 1
    print("Total No of Similarities: " + str(similarity_counter))

if __name__ == "__main__":
    main()
```

```python
# Function Definitions
# For Q B
def split_words(inp_quote):
    """
    The function takes the input as a string.
    First, it lower cases the string and then splits it based on spaces.
    Then, to remove any empty strings or '' that might've gotten detected as words,
    it does a simple join-split based on space ' ' to remove those empty strings and returns a list
    of words
    
    Parameters
    --------------------------------------------------------------------------
    inp_quote : str
    The full quote string that needs to be split into constituent words
    
    
    Returns
    --------------------------------------------------------------------------
    word_list: list of str
    List of words that constitute the inputted string

    """
    inp_quote = inp_quote.lower()   #Let's lower case it
    inp_quote = re.sub('[\'`]', '', inp_quote)  # Let's remove those pesky apostrophes
    word_list = re.split('[\s\W]', inp_quote)  # Split based on white spaces and don't select any non numeric characters (except underscore)
    word_list = ' '.join(word_list).split()   # remove any blank '' entries from list
    return word_list



# For Q C
def getWordCountDict(inp_quote):
    """
    The function takes a string (a full quote) as its inpuit.
    First, it splits the string into its constituent words using the split_words fucntion.
    Initialize a word_set which will store words that have already been processed/counted.
    Iterate through each word in the list of words.
    If word has already been counted i.e if it exists in word_set, move to next word.
    If word does not exist in word_set, perform following -
    1) Add current word to word_set and count instances of word
    in the entire word list.
    2)Add new element in word_dict with current word as key and word count as value
    3) Continue to next word
    
    Perform the following for each word in word_list
    Return dictionary with unique words as keys and their counts as values.
    
    Parameters
    --------------------------------------------------------------------------
    inp_quote : str
    The full quote string
    
    
    Returns
    --------------------------------------------------------------------------
    word_dict: dictionary
    Dictionary where values are unique words of inpuit_quote and values are their respective counts

    """
    word_list = split_words(inp_quote)   # Leverage previous function to get list of words
    word_set = ['%']   # Initialize word set and word dict 
    word_dict ={}
    for word in word_list:
        if word in word_set:   # If word is already counted, don't bother again
            continue
        else:     # If word is not yet counted (not in word set)
            word_set.append(word)   # Add word to word set
            count = word_list.count(word)     # count occurrences of word in the word list
            word_dict[word] = count           # add element to word count dictionary
    return word_dict       # Return word list dictionary
```

```python
# Q C - Post-Listings dictionary
# Let's create a post listings dictionary

post_listings_dictionary ={}    # Initialize the dictionary

for quote in quotes:
    # Iterate over every quote
    post_listing = getWordCountDict(quote)   # Refer the 2nd cell with function definitions
    post_listings_dictionary[quote] = post_listing

print(post_listings_dictionary[quotes[0]])

# Q C done
```

```python
# Q D Reverse Post Listings Dictionary
# Let's build this as we iterate through every quote

reverse_post_listings_dictionary = {}    # Initialize the reverse post listings dictionary

for quote in quotes:    # Iterate through every quote
    word_list = split_words(quote)  # Get a list of words in the quote
    word_set=[]   # Initialize word set
    for word in word_list:
        if word in word_set:   # If word is already counted, don't bother again
            continue
        else:
            word_set.append(word)   # Add word to word set
            count = word_list.count(word)   # Get count of word occurrence in quote
            if word in list(reverse_post_listings_dictionary.keys()):    # If word already exists in reverse dict, update its value dictionary
                reverse_post_listings_dictionary[word][quote] = count    # Update value dictionary
            else:
                reverse_post_listings_dictionary[word] = {quote:count}   # If word doesn't exists yet, initialize the value dictionary
                
print(reverse_post_listings_dictionary['entertainer'])

# Q D Done
```

```python
# Q E TF-IDF Function
# Let's define the TF-IDF function

def tfIDF(word, quote):
    """
    The function calculates the TF-IDF value for a given word and a quote.
    First, make some basic sanity checks -
    1) The quote must exist in full quotes list
    2) The word entered must exist in quote
    
    Now, get count of word that occurs maximum no. of times in the input quote (use key-values
    from the post liostings dictionary)
    Get count of how many times the input word exists in the input quote (using post listings
    dictionary)
    Calculate TF = word count / max word count in quote
    
    Nopw for IDF,
    Get total no. of full-quotes
    Get count of quotes in which this word appears (total no. of values that exist for the word(key) in the
    reverse post listings dictionary)
    
    Calculate IDF = log(total no. of full-quotes/count of quotes in which this word appears)
    
    Calculate TFIDF = TF*ODF and return value
    
    
    Parameters
    --------------------------------------------------------------------------
    word : str
    A word that might exist in a quote
    
    quote: str
    Corresponding quote for which the TF and IDF needs to be calculated
    
    
    Returns
    --------------------------------------------------------------------------
    td*idf: float
    The TDIDF value calculated
    """
    # Basic sanity check - 
    # If quote does not exist, return -1 with a message
    # If word does not exist, just return 0 since TF will be 0 and hence TF*IDF will be 0, no need to calculate
    
    if quote not in quotes:
        print("Entered quote does not exist. Please check input")
        return -1
    elif word not in list(post_listings_dictionary[quote].keys()):
        print("Entered word does not exist in entered quote")
        return 0
    
    # If word exists
    # Calcuate TF
    max_count = max(list(post_listings_dictionary[quote].values()))
    word_count = post_listings_dictionary[quote][word]
    tf = word_count/max_count
    
    # Calculate IDF
    N = len(quotes)   # Total quotes
    quote_freq = len(list(reverse_post_listings_dictionary[word].keys()))   # Total quotes in which the word appears
    idf = math.log(N/quote_freq)  
    
    return tf*idf   # Calculate and return TF-IDF


tfidf = tfIDF('entertainer', "An actor is at most a poet and at least an entertainer. - Marlon Brando")   # Enter any input here
print(tfidf)

# Q E Done
```

```python
# Q F Quote searching single word
# Let's define a search function

def getQuotesIDF (word):
    """
    The function returns a dictionary where keys are the full quotes in which the input words
    occurs and the values are the corresponding TFIDF values of the input word and the quote.
    
    First, convert input word to lower case since all words in post listing and reverse post listings
    dictionary are in lower case.
    Perform basic input checking -
    1) Check if word exists in reverse post listings dictionary.
    
    Create a copy of the reverse post listings dictionary value dictioanry for the input word.
    Iterate through every quote in the quote:value dictionary for input word from its reverse
    post listings dictionary value.
    Calculate TF-IDF for each input word-quote pair and update the count value with the new TFIDF value
    for that quote in the reverse post listings dictionary copy.
    
    Return the fully updated quote-idf dictionary with quotes where the word appeared as key and 
    corresponding TFIDF score as value.
    
    Parameters
    --------------------------------------------------------------------------
    word : str
    A word that might exist in a quote

    
    Returns
    --------------------------------------------------------------------------
    quote_tfidf: dict
    Dictionary with quotes where the word appeared as key and corresponding TFIDF score as value.
    
    
    """
    # First, make the word into lower case
    word = word.lower()
    # Basic sanity check -
    # Check if word exists in reverse post listings dictionary
    if word not in list(reverse_post_listings_dictionary.keys()):
        print("Word does not exist in quote lexicon.")
        return -1
    # Get reverse postings list dictionary as a start point
    quote_tfidf = reverse_post_listings_dictionary[word]
    
    # Now for every quote, calculate TFIDF and update the values against quotes as TFIDF values instead of word counts
    for quote in reverse_post_listings_dictionary[word]:
        # Calculate TFIDF
        tfidf = tfIDF(word, quote)
        quote_tfidf[quote] = tfidf
    return quote_tfidf

print (getQuotesIDF('dream'))

# Q F Done
```

```python
# Q G Quote search using multiple words

def getQuotesIDFMulti(word_list):
    """
    Function takes a list of words (strings) as input and returns a combined quote-TFIDF dictionary
    as output.
    Convert all words  to lower case.
    Perform basic input checks -
    1) Make sure each word exists in reverse post listings dictionary.
    2) If word doesn't exist, remove the word and let user know.
    3) If all words end up being removed, let user know to enter at least 1 valid word.
    Initialize output dictionary
    Iterate through every word in input word list -
    1) Use getQuotesIDF function to get TFIDF dictionary where quotes where the word appears as key and
    corresponding TFIDF score as value.
    2) Iterate through every qoute in the TFIDF dictionary -
        1) If quote already exists as key in output dictionary (this would be the case when a word in input 
        word list that has already been processed has also appeared in the same quote), update the TFIDF 
        value associated with quote by adding the current TFIDF value to the existing TFIDF value against 
        the quote in the output dictionary.
        2) If quote does not exist in output dictionary, add it to output dictionary with quote as key and
        TFIDF score as value
        
    After processing all words, return final combined TFIDF output dictionary.
    
    Parameters
    --------------------------------------------------------------------------
    word_list : list of str
    A list of multiple (possible) words that might exist in a quote

    
    Returns
    --------------------------------------------------------------------------
    quote_tfidf_multi: dict
    Combined dictionary with quotes where the word appeared as key and corresponding TFIDF score 
    as value. If multiple words exist in same quote, corresponding TFIDF score (value) against the
    quote will be addition of the TFIDF scores of both words.
    
    """
    # Basic sanity check -
    # Check if word(s) exists in reverse post listings dictionary
    for word in word_list:
        word = word.lower()   # make word in lower case
        if word not in list(reverse_post_listings_dictionary.keys()):
            print("Word ''",word,"'' does not exist in quote lexicon.")
            word_list.remove(word)   # Remove those words that don't exist in reverse post listings dictionary
    if  not word_list:
        print("None of the words entered exist in quote lexicon.")
        return -1
    
    # Get individual dictionaries of each word
    quote_tfidf_multi ={}
    for word in word_list:
        word_idf = getQuotesIDF(word)
        for quote in word_idf:
            if quote in list(quote_tfidf_multi.keys()):
                quote_tfidf_multi[quote] += word_idf[quote]
            else:
                quote_tfidf_multi[quote] = word_idf[quote]
                
    return quote_tfidf_multi


# Look at the first quote of below two -
# Quoting these out because Jupyter cuts out the output after the first two prints statements
#print(getQuotesIDFMulti(['dream']),"/n/n/n")  # IDF = 2.353
#print(getQuotesIDFMulti(['america']),"\n\n\n") # IDF = 2.844
# Now look at first quote of output -

print(getQuotesIDFMulti(['dream', 'america'])) # IDF = 2.353 + 2.844 = 5.197

# Q G Done
```
